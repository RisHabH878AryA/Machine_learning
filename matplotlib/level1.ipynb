```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("VIT_Bhopal_Update_26DEC25.xlsx - List.csv")
```

```python
df["id"] = df.index + 1
df["Completion Percentage"] = df["Completion Percentage"].str.replace("%", "").fillna(0).astype(int)
comp_percent = df["Completion Percentage"]
student = df["id"]
```

```python
plt.plot(student,comp_percent)
```

```python
plt.scatter(student,comp_percent)
```

```python
degree = df["Degree"]
plt.bar(degree, comp_percent)
```

```python
plt.hist(comp_percent, bins=20)
plt.show()
```
