```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("VIT_Bhopal_Update_26DEC25.xlsx - List.csv")
df["id"] = df.index + 1
df["Completion Percentage"] = df["Completion Percentage"].str.replace("%", "").fillna(0).astype(int)
df
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df["id"], df["Completion Percentage"], alpha=0.1, label="Meow")
plt.title("Completion percentage")
plt.xlabel("Number of students")
plt.ylabel("Completion Percentage")
plt.grid(True)
plt.legend()
plt.show()
```

```python
btech = df["id"][df.Degree == "B.Tech"]
mtech = df["id"][df.Degree == "M.Tech 5 yrs integrated"]

plt.figure()
plt.scatter(btech, df["Completion Percentage"][df.Degree == "B.Tech"], label="B.Tech")
plt.scatter(mtech, df["Completion Percentage"][df.Degree == "M.Tech 5 yrs integrated"], label="M.Tech")

plt.xlabel("ID")
plt.ylabel("Completion Percentage")
plt.legend()
plt.grid(True)
plt.show()
```
