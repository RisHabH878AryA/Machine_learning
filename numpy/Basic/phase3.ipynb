```python
# imports and stuff
import numpy as np
from machine_learning.utility import *
import matplotlib.pyplot as plt
```

### Hands-on-practice

#### Data generation

```python
restaurant_id = np.array([i for i in range(1,6)])
_i2021 = np.random.choice([i*5000 for i in range(1,55)], size=(5))
_i2022 = np.random.choice([i*5000 for i in range(1,55)], size=(5))
_i2023 = np.random.choice([i*5000 for i in range(1,55)], size=(5))
_i2024 = np.random.choice([i*5000 for i in range(1,55)], size=(5))

# sales_list = [_i2021, _i2022, _i2023, _i2024]

# sales_list = [np.array(x, dtype="f") for x in sales_list]

# _i2021, _i2022, _i2023, _i2024 = sales_list

    
sales_data = np.vstack((restaurant_id,_i2021,_i2022, _i2023, _i2024))

show(
    restaurant_id = restaurant_id,
    _i2021=_i2021,
    _i2022=_i2022,
    _i2023=_i2023,
    _i2024=_i2024,
    # sales_data=sales_data
)
sales_data = sales_data.transpose()
sales_data
```

#### Data info

```python
# Shape, size, dimension, datatype etc
show(
    data = sales_data,
    data_shape = sales_data.shape,
    data_size = sales_data.size,
    dimensions_of_data = sales_data.ndim,
    data_datatype = sales_data.dtype
)
```

#### calculations

```python
# Sum, min, max, mean, cumulative mean 

total_sales_per_restaurant = np.sum(sales_data[:,1:], axis=1)
total_sales_per_year = np.sum(sales_data[:, 1:], axis=0)
average_sales_each_year = np.mean(sales_data[:,1:], axis=0)
average_sales_per_restaurant = np.mean(sales_data[:,1:], axis=1)
max_sales_from_each_restaurant = np.max(sales_data[:,1:], axis=1)
min_sales_each_year = np.min(sales_data[:,1:], axis=0)

cumulative_sum_per_restaurant = np.cumsum(sales_data[:,1:], axis=1)
cumulative_mean_per_restaurant = np.mean(cumulative_sum_per_restaurant, axis=1)
cumulative_mean_yearly = np.mean(cumulative_sum_per_restaurant, axis=0)

show(
total_sales_per_restaurant=total_sales_per_restaurant,
total_sales_per_year=total_sales_per_year,
average_sales_each_year=average_sales_each_year,
average_sales_per_restaurant=average_sales_per_restaurant,
max_sales_from_each_restaurant=max_sales_from_each_restaurant,
min_sales_each_year=min_sales_each_year,
cumulative_sum_per_restaurant=cumulative_sum_per_restaurant,
cumulative_mean_per_restaurant=cumulative_mean_per_restaurant,
cumulative_mean_yearly=cumulative_mean_yearly
)
```

#### Plotting

```python
years = np.arange(2021, 2025)

plt.figure(figsize=(10,7))
plt.plot(years, cumulative_mean_yearly, marker='o')
plt.title("Average yearly cumulative sales across all restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
