### Images and Storing arrays

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
from machine_learning.utility import show
```

#### Saving and loading arrays

Saving arrays as npy extension

```python
arr1 = np.arange(1,19,2).reshape(3,3)
arr2 = np.random.rand(1,10)
arr3 = np.array([(x,y) for x in range(2) for y in range(10,5,-1)])

np.save(file="Array1",arr= arr1)
np.save(file="Array2",arr= arr2)
np.save(file="Array3",arr= arr3)
```

Loading saved arrays

```python
np.load(file="Array1.npy")
```

#### Image by array

```python
logo = np.load(file="numpy_logo.npy")
plt.figure(figsize=(10,5))

plt.subplot(121)
plt.imshow(logo)
plt.title("Numpy logo.. hehe!")

plt.subplot(122)
plt.imshow(1 - logo)
plt.title("Numpy negative logo.. \nmore hehe!")

plt.grid(False)
plt.show()

show(
    image_to_array = logo
)
```
