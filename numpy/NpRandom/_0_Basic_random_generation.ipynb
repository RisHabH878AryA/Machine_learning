### Random numbers generation

```python
# imports and stuff
import numpy as np
from machine_learning.utility import show
```

random integers and floats

```python
show(
    random_int_from_0_to_100 = np.random.randint(100),
    random_float = np.random.rand(),
)
```

Random array of int and float

```python
show(
    random_int_vector = np.random.randint(100, size=(3)),
    random_int_matrix = np.random.randint(100, size=(3,3)),
    
    random_float_vector = np.random.rand(5),
    random_float_matrix = np.random.rand(3,3)
)
```

Random number from Array

```python
show(
    random_choice = np.random.choice([34,7,4,5,34]),
    randomly_choosen_numbered_array = np.random.choice([69,6,7,420], size=(3,3)),
)
```
