### Data distribution, Permutation & Seaborn

```python
# imports and stuff
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
from machine_learning.utility import show
```

#### Random Distribution

```python
show(
    dist1d = random.choice([1,3,5,7,9], p=[0.4,0.15,0.2,0.15,0.1], size=(100)),
    dist2d = random.choice([1,3,5,7,9], p=[0.4,0.15,0.2,0.15,0.1], size=(3,3))
)
```

#### Random Permutations

```python
arr1 = np.arange(7)
show(
    original = arr1,
    permutation = random.permutation(arr1)    
     )

arr2 = np.arange(10,3,-1)
random.shuffle(arr2)
show(
    shuffled = arr2,
)
```

#### Seaborn

```python
array = np.arange(100)
sns.displot(array, kind='kde')
show(array = array)
```
