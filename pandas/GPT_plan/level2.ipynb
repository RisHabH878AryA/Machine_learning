```python
# Duplicates
df.duplicated().sum()
df.drop_duplicates(inplace=False)
df.drop_duplicates(['Name'])
# df.loc[len(df)] = {"Name": "Jeff Withey"}
df
```

```python
# Renaming columns

# all
# new_col_names = []
# for x in list(df.columns):
#     new_col_names.append(x.lower())
# new_col_names

# space -> underscore
df.columns = [col.lower().replace(" ", "_") for col in df.columns]
df

# specifically one
# df.rename({'team':'Team'}, axis=1, inplace=True)
# df.rename(columns={'number': 'Number'}, inplace=True)
# df
```

```python
# Filtering and querying
df[df['age'] > 20]

df[df['full_name'] == 'James Young']

mask1 = (df['weight'] < 180)
mask2 = (df['weight'] > 170)
df[mask1 & mask2]

df[df["full_name"].str.startswith('Ar')]

df.query("full_name.str.startswith('A')")

df.query("number*6 > weight")
```

```python
# Sorting
df.sort_values('age', ascending=False)
df.sort_values([ 'weight', 'age'])
df.sort_values(by=[ 'age', 'weight'], ascending=[True, False])
```

```python
# Aggregation

# mean, median, mode, max, min
df['weight'].mean()
df['weight'].median()
df['weight'].mode()
df['weight'].max()
df['weight'].min()

df['weight'].agg(["mean", "median", "max", "min"])
df['weight'].describe()

# group by 
df.groupby('college')['weight'].agg(["count", "mean"])
df.groupby(['college','team'])['weight'].agg(["count", "mean"])
```

```python
# Basic column chatak chala
l = (df["full_name"].str.split(" ").to_list())
fname = [x[0] for x in l]
lname = [x[-1] for x in l]
df['first_name'] = fname
df['last_name'] = lname
df.head()

df["is_adult"] = df['age'] > 18
```

```python
# Data type conversion
# int
# df["age"] = df["age"].astype(int)
# category
print((df['team']).dtype)
df["team"] = df["team"].astype("category")
print((df['team']).dtype)
df

for x in df.columns:
    print(df[x].name, " ||| " ,  df[x].dtype)
```
