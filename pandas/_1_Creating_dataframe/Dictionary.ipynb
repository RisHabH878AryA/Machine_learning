 Creating Dataframe using dictionary

```python
import pandas as pd
```

```python
# Simple dictionary
df = pd.DataFrame({"Name": ["Rishabh", "Aryan", "Meow"], "Number" : [1,2,3]})
df.head()
```

```python
# Dictionary of series
winners = pd.Series(["Participant1","Participant3","Participant2"])
positions = pd.Series([1,2,3])
df = pd.DataFrame({
    "Winners": winners,
    "Positions": positions})
df.head()
```
