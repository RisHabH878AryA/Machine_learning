Creating dataframe from lists

```python
import pandas as pd
import random
```

```python
# Dictionary of lists
name = ["name" + str(x) for x in range(10)]
age = [float(random.random()*x) for x in range(100,110)]
roll_no = [x for x in range(1000,1010)]
diction = {"name": name, "age" : age, "roll_no":roll_no}
df = pd.DataFrame(diction)
df.head()
```

```python
# using zip function
nums = ["one", "ten", "hundred", "thousand"]
values = [1,10,100,1000]
df = pd.DataFrame(list(zip(nums, values)), columns=["Names", "Values"])
print(df)
```

```python
# Multi-dimensional list
data = [["Name", "Rishabh"], ["Class", "10th"], ["RollNumber", 26], ["Section", 'C']]
df = pd.DataFrame(data, columns=["label", "data"])
df.head()
```

```python
# Changing datatype after creating Dataframe
randomDigits = [random.random()*x for x in range(100,110)]
df = pd.DataFrame(randomDigits, columns=["Random_Digits"])
df.head()
df["Random_Digits"] = df["Random_Digits"].astype(int)
df.head()
```

```python
# Using index and column names
data = ["Aparna", "Pankaj", "Sudhir", "Garvit"]
df = pd.DataFrame(data, index=range(1000,1004), columns=["Names"])
df.head()
```
