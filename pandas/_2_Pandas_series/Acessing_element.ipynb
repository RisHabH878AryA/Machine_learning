Two ways to access elements:
1] position based indexing
2] Label based indexing

```python
import pandas as pd
import numpy as np
```

```python
# 1] position based:
ser = pd.Series(np.random.randn(5))
print(ser[:5])
```

```python
# 2] Label based indexing
np_array = np.array([x for x in range(1000, 1026)])
# print(np_array)
ser = pd.Series(
    np_array,
    index=[chr(ord('a') + x) for x in range(26)]
)

print(ser[['m','e','o','w']])
```

```python
# 2 
ser = pd.Series(np.arange(3,9), index=[chr(ord('a') + x) for x in range(6)])
print(ser['a':'d'])
```
