Indexing a series using .loc[]

```python
import pandas as pd
```

1. Indexing with .loc[]
-->using loc : retrives elements with labels

```python
# Selecting a single row by label
data = pd.read_csv("nba.csv", index_col='Name')

row = data.loc["Avery Bradley"]
print(row)
```

```python
# Selecting multiple rows by label
row = data.loc[["Avery Bradley", "R.J. Hunter"]]
row
```

```python
# using iloc: retrieve data by position
# (loc but only uses integer location to make its selections)
df = pd.read_csv("nba.csv")
ser = pd.Series(df['Name'])
ser.iloc[3:6]
```
