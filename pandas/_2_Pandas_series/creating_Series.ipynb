```python
import pandas as pd
import numpy as np
import random
```

```python
# Empty panda series
ser = pd.Series()
print(ser)
```

```python
# Creating panda series from list
data = [random.random()*x for x in range(10, 15)]
ser = pd.Series(data)
print(ser)
```

```python
# From numpy arrays
data = np.array(['a', 'b', 'c', 'f'])
ser = pd.Series(data, index=['i','ii','iii','iv'])
print(ser)
```

```python
# From dictionary
data = {'Name': 'Rishabh', 'Class': '10th', 'Rollno':1000, 'Section': 'C'}
ser = pd.Series(data)
print(ser)
```

```python
# from numpy functions
ser1 = pd.Series(np.linspace(1,10, 6))
ser2 = pd.Series(np.random.randn(5))
print("ser1:\n", ser1)
print("ser2:\n", ser2)
```

```python
# using list comprehension
ser = pd.Series(range(1,20,5), index=[x for x in 'meow'])
print(ser)
```
