```python
# Pandas Dataframe
## -> It is a two dimensional table like structure in python.
## -> where data is arranged in rows and columns.
## -> It consists of DATA, ROWS, COLUMNS.
```

```python
# Creating pandas dataframe:
# Dataframes can be created directly from 
# -- python objects like lists and dictionaries or by reading data from external files like CSV, Excel, SQL databases
```

Creating Dataframes

```python
import pandas as pd
import numpy as np
```

```python
# using a list
list1 = [[1,2,3,4,5], [5,4,3,2,1]]
df = pd.DataFrame(list1)
print(df)

# using dictionaries
dict1 = {"Name": [0,1,2], "roll_no":[1,2,3]}
df = pd.DataFrame(dict1)
print(df)

# using list of dictionaries
students = [
    {"Name" : "A", "Age": 20},
    {"Name": "B", "Age" : 21}
]
df = pd.DataFrame(students)
print(df)

#using dict of numpy array
data = np.array([[1,1], [2,2]])
df = pd.DataFrame(data, columns=['A', 'B'])
print(df)
```
