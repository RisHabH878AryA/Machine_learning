Operations in pandas

```python
import pandas as pd
```

```python
# Loading data
df = pd.read_csv("people_data.csv")
print(df.head()) # outputs first few (5) rows of the dataframe
```

```python
print(df.info()) # metadata of the dataframe
```

```python
# Handling missing data
print(df.isnull().sum)
df = df.fillna(0)  # replaces missing values with 0
```

```python
# Selecting and filtering data
ages = df[df['Age'] > 25]
print(ages)
```

```python
# Adding and removing columns
df["Real Age"] = 2025 - df["Date of birth"].str.split("-").str[0].astype(int)
```

```python
df.head()
```

```python
# Grouping data
response = df.groupby('First Name')['Real Age'].sum()
print(response)
```
